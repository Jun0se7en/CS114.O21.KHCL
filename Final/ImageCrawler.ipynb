```python
from icrawler.builtin import GoogleImageCrawler
import time
```

```python
yamaha = ['luvias', 'exciter', 'janus', 'grande', 'sirius']

# Yamaha
count = 100
yamaha_crawler = GoogleImageCrawler(feeder_threads=1, parser_threads=2, downloader_threads=4, storage={'root_dir': './yamaha'})
yamaha_crawler.crawl(keyword='yamaha_motorcycle', max_num = 100)
for i in yamaha:
    yamaha_crawler.crawl(keyword=f'yamaha_motorcycle_{i}', max_num = 50, file_idx_offset=count)
    count += 50
```
